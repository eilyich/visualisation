–§—É–Ω–∫—Ü–∏–∏ –∏ —Ü–∏–∫–ª—ã –æ–±–ª–µ–≥—á–∞—é—Ç –∂–∏–∑–Ω—å –≤–æ –º–Ω–æ–≥–∏—Ö –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∑–∞–¥–∞—á–∞—Ö, –≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤. –í —ç—Ç–æ–π —à–ø–∞—Ä–≥–∞–ª–∫–µ —è —Å–æ–±—Ä–∞–ª –±–∞–∑–æ–≤—ã–π –º–∏–Ω–∏–º—É–º —Ç–æ–≥–æ, —á—Ç–æ –ø–æ–º–æ–∂–µ—Ç –æ—Ç—Ä–∏—Å–æ–≤–∞—Ç—å –≤ –Ω–æ—É—Ç–±—É–∫–µ –ª—é–±–æ–π –≥—Ä–∞—Ñ–∏–∫, –∫–∞–∫ –ø—Ä–æ—Å—Ç–æ–π —Ç–∞–∫ –∏ –±–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã–π - –ø—Ä–∏–Ω—Ü–∏–ø—ã –≤–µ–∑–¥–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ.    

–Ø –±—É–¥—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±–∏–±–ª–∏–æ—Ç–µ–∫—É `seaborn`.

–í –Ω–∞—á–∞–ª–µ –∫–∞–∫ –æ–±—ã—á–Ω–æ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale = 1.5, style="darkgrid")
```

üêß

```python
data = sns.load_dataset("penguins").dropna()
display(data.head())
display(data.describe())
```

–î–ª—è –Ω–∞—á–∞–ª–∞ –Ω–∞—Ä–∏—Å—É–µ–º –ø—Ä–æ—Å—Ç–æ–π –≥—Ä–∞—Ñ–∏–∫, —Å–∫–∞–∂–µ–º —Ä–∞—Å—Å–µ—è–Ω–∏—è, —á–µ—Ä–µ–∑ plt.subplots - –ø–æ–∫–∞ –≥—Ä–∞—Ñ–∏–∫ –≤—Å–µ–≥–æ –æ–¥–∏–Ω, –Ω–æ —ç—Ç–æ—Ç –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ç–¥–µ–ª—å–Ω–æ –∑–∞–¥–∞—Ç—å, —Å–≤–æ–µ–≥–æ —Ä–æ–¥–∞, —Ä–∞–º–∫—É –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞.

```python
fig, ax = plt.subplots(figsize=(18, 8))                                # —Å–æ–∑–¥–∞–Ω–∏–µ "—Ä–∞–º–∫–∏"
ax = sns.scatterplot(data=data, x="bill_length_mm", y="bill_depth_mm") # —Å–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞

plt.show()
```

–ï—Å–ª–∏ –∫–æ–ª–æ–Ω–æ–∫ –≤ —Ç–∞–±–ª–∏—Ü–µ –º–Ω–æ–≥–æ, –∏ –Ω—É–∂–Ω–æ –±—ã—Å—Ç—Ä–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤–æ –≤—Å–µ—Ö, –ø—Ä–æ—â–µ –≤—Å–µ–≥–æ —ç—Ç–æ —Å–¥–µ–ª–∞—Ç—å –≤ —Ü–∏–∫–ª–µ. –ì—Ä–∞—Ñ–∏–∫–∏ –æ—Ç—Ä–∏—Å—É—é—Ç—Å—è –ø–æ–æ—á–µ—Ä–µ–¥–∏.

```python
for col in data.columns:                     # –¥–ª—è '–∫–∞–∂–¥–æ–π_–∫–æ–ª–æ–Ω–∫–∏_–ø–æ_–æ—á–µ—Ä–µ–¥–∏' –≤ '—Ç–∞–±–ª–∏—Ü–∞.–≤—Å–µ_–∫–æ–ª–æ–Ω–∫–∏'
    fig, ax = plt.subplots(figsize=(6, 4))
    ax = sns.histplot(data=data, x=col)      # .histplot(—Ç–∞–±–ª–∏—Ü–∞, –∫–∞–∂–¥–∞—è_–∫–æ–ª–æ–Ω–∫–∞_–ø–æ_–æ—á–µ—Ä–µ–¥–∏)

plt.show()
```

–ù–æ –ª–∏—á–Ω–æ –º–Ω–µ –≤—Å–µ–≥–¥–∞ –Ω–µ –Ω—Ä–∞–≤–∏–ª–∏—Å—å —É—Ö–æ–¥—è—â–∏–µ –∫—É–¥–∞-—Ç–æ –¥–∞–ª–µ–∫–æ –≤–Ω–∏–∑ "–ø—Ä–æ—Å—Ç—ã–Ω–∏" –æ—Ç—Ä–∏—Å–æ–≤–∞–Ω–Ω—ã—Ö –≤ —Ü–∏–∫–ª–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤.
–î–ª—è –ª—é–±–∏—Ç–µ–ª–µ–π –∏–∑—è—â–Ω–æ–≥–æ –∏ –ø—Ä–∏–¥—É–º–∞–Ω—ã —Å–∞–±–ø–ª–æ—Ç—ã. –° –∏—Ö –ø–æ–º–æ—â—å—é –º–æ–∂–Ω–æ —Ä–∞—Å–ø–æ–ª–æ–∂–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ —à–∏—Ä–∏–Ω—É –∏ –≤—ã—Å–æ—Ç—É, —Ö–æ—Ç—å 2, —Ö–æ—Ç—å 5, —Ö–æ—Ç—å 10.   

–ö–ª—é—á–µ–≤–æ–π –º–æ–º–µ–Ω—Ç, —á—Ç–æ —Ç–µ–ø–µ—Ä—å –≤ `plt.subplots` –º—ã –±—É–¥–µ–º –∑–∞–¥–∞–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ (–≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã—Ö —Ä—è–¥–æ–≤) –∏ –∫–æ–ª–æ–Ω–æ–∫ (–≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã—Ö):   
`fig, ax = plt.subplots(–≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å, –≤–µ—Ä—Ç–∏–∫–∞–ª—å)`
–ê –ø—Ä–∏ –æ—Ç—Ä–∏—Å–æ–≤–∫–µ –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ —É–∫–∞–∑—ã–≤–∞—Ç—å –≤ –∫–∞–∫–æ–π —è—á–µ–π–∫–µ –æ–Ω –ª–µ–∂–∏—Ç - –Ω–∞–ø—Ä–∏–º–µ—Ä `ax[1, 3]` –±—É–¥–µ—Ç –∑–Ω–∞—á–∏—Ç—å **–≤—Ç–æ—Ä–∞—è** —Å—Ç—Ä–æ–∫–∞, **—á–µ—Ç–≤–µ—Ä—Ç–∞—è** –∫–æ–ª–æ–Ω–∫–∞.     
! –æ–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ - –≤ `plt.subplots(...)` - —á–∏—Å–ª–∞ –æ–±–æ–∑–Ω–∞—á–∞—é—Ç **–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ** —Å—Ç—Ä–æ–∫ –∏ –∫–æ–ª–æ–Ω–æ–∫. –ê –ø—Ä–∏ —É–∫–∞–∑–∞–Ω–∏–∏ —è—á–µ–µ–∫ —Ç—ã –æ–±—Ä–∞—â–∞–µ—à—å—Å—è **–∫ –∏–Ω–¥–µ–∫—Å–∞–º**, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞—á–∏–Ω–∞—é—Ç—Å—è —Å 0, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –¥–ª—è —Ä–∞–º–∫–∏ `plt.subplots(3, 4)` –º–∞–∫–∏–º–∞–ª—å–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã –±—É–¥—É—Ç `ax[2, 3]`    
–ö–æ–Ω–µ—á–Ω–æ, –ø—Ä–æ—â–µ –≤—Å—ë —ç—Ç–æ –ø–æ–Ω—è—Ç—å –≤ —Å–∞–º–æ–º –∫–æ–¥–µ, –∏ —Å–µ–π—á–∞—Å –æ–Ω –±—É–¥–µ—Ç.    

–¢–∞–∫ –∫–∞–∫ –∫–∞–∂–¥—ã–π –≥—Ä–∞—Ñ–∏–∫ –æ—Ç—Ä–∏—Å–æ–≤—ã–≤–∞–µ—Ç—Å—è –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ (–Ω–æ–≤–æ–π, –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–æ–π), –º–æ–∂–Ω–æ —Ä–∏—Å–æ–≤–∞—Ç—å —Ä—è–¥–æ–º –ª—é–±—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏, –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. –ù–∞–ø—Ä–∏–º–µ—Ä —Ç–∞–∫:

```python
fig, ax = plt.subplots(2, 2, figsize=(18, 12))                         # –∑–∞–¥–∞–ª–∏ 2 —Å—Ç—Ä–æ–∫–∏ –∏ 2 –∫–æ–ª–æ–Ω–∫–∏
sns.histplot(data=data, x='bill_length_mm', ax=ax[0, 0])               # —Å—Ç—Ä–æ–∫–∞ 0, –∫–æ–ª–æ–Ω–∫–∞ 0
sns.kdeplot(data=data, x='bill_length_mm', hue='species', ax=ax[0, 1]) # —Å—Ç—Ä–æ–∫–∞ 0, –∫–æ–ª–æ–Ω–∫–∞ 1
sns.histplot(data=data, x='body_mass_g', ax=ax[1, 0])                  # —Å—Ç—Ä–æ–∫–∞ 1, –∫–æ–ª–æ–Ω–∫–∞ 0
sns.kdeplot(data=data, x='body_mass_g', hue='sex', ax=ax[1, 1])        # —Å—Ç—Ä–æ–∫–∞ 1, –∫–æ–ª–æ–Ω–∫–∞ 1

plt.show()
```

–ú–æ–∂–Ω–æ –∏ —Å–ª–æ–∂–Ω—ã–π —Å–∞–±–ø–ª–æ—Ç –æ—Ç—Ä–∏—Å–æ–≤–∞—Ç—å –≤ —Ü–∏–∫–ª–µ. –ù–æ –¥–ª—è —ç—Ç–æ–≥–æ –Ω–∞–º –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è –Ω–∞–±–æ—Ä –∏–Ω–¥–µ–∫—Å–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç –¥–≤–∞ –∑–Ω–∞—á–µ–Ω–∏—è - —Å—Ç—Ä–æ–∫—É –∏ –∫–æ–ª–æ–Ω–∫—É. –ß—É—Ç—å –Ω–∏–∂–µ –º—ã –∑–∞–¥–∞–¥–∏–º —Å–∞–±–ø–ª–æ—Ç —Ä–∞–∑–º–µ—Ä–æ–º 3—Ö2 (3 —Å—Ç—Ä–æ–∫–∏ 2 –∫–æ–ª–æ–Ω–∫–∏), –∞ —Å–µ–π—á–∞—Å —Å–≥–µ–Ω–∏—Ä–∏—Ä—É–µ–º –∏–Ω–¥–µ–∫—Å—ã, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ–º –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å –ø–æ—Ç–æ–º –∫–∞–∂–¥–æ–º—É –≥—Ä–∞—Ñ–∏–∫—É –ø–æ –æ—á–µ—Ä–µ–¥–∏.     


–î–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–æ–∂–µ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ü–∏–∫–ª–æ–º, –µ—â—ë –æ–¥–Ω–∞ —Ö–æ—Ä–æ—à–∞—è –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–≤—Ç–æ—Ä–∏—Ç—å –º–∞—Ç–µ—Ä–∏–∞–ª. 

```python
grid = []                     # —Å–æ–∑–¥–∞—ë–º –ø—É—Å—Ç—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è '—Å–µ—Ç–∫–∏'
for i in range(3):            # –¥–ª—è _–∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏_ –∏–∑ _—Ç—Ä—ë—Ö_
    for j in range(2):        # –≤ —ç—Ç–∏—Ö —Å—Ç—Ä–æ–∫–∞—Ö –¥–ª—è _–∫–∞–∂–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞_ –∏–∑ _–¥–≤—É—Ö_
        grid += [(i, j)]      # –¥–æ–±–∞–≤–ª—è–µ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é _–∏–Ω–¥–µ–∫—Å –ø–æ —Å—Ç—Ä–æ–∫–µ_, _–∏–Ω–¥–µ–∫—Å –ø–æ —Å—Ç–æ–ª–±—Ü—É_
grid
```

–ü–æ–ª—É—á–∏–ª—Å—è —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π. –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ª—É—á—à–µ –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –∫–æ—Ä—Ç–µ–∂–∏ (*tuple, –≤–æ—Ç —ç—Ç–æ –≤–æ—Ç –≤ –∫—Ä—É–≥–ª—ã—Ö —Å–∫–æ–±–∫–∞—Ö*).   


–ú–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å —ç—Ç–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≥–æ—Ä–∞–∑–¥–æ –∫–æ—Ä–æ—á–µ, –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É. –ï–≥–æ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –≤ —Ç–æ–º, —á—Ç–æ –µ–≥–æ –º–æ–∂–Ω–æ –Ω–µ –ø—Ä–∏—Å–≤–∞–∏–≤–∞—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π (–∫–∞–∫ –º—ã –¥–µ–ª–∞–µ–º —Å–µ–π—á–∞—Å), –∞ —Å—Ä–∞–∑—É –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º –≤ —Ñ—É–Ω–∫—Ü–∏—é. 

```python
grid =  [(i, j) for i in range(3) for j in range(2)]
grid
```

–ü–æ–ª—É—á–∏–ª–æ—Å—å —Ç–æ –∂–µ —Å–∞–º–æ–µ. –°–µ—Ç–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤ —Ü–∏–∫–ª–µ –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å—Å—è –ø—Ä–∏ –æ—Ç—Ä–∏—Å–æ–≤–∫–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤, –≥–æ—Ç–æ–≤–∞.

```python
data_plot_columns = data.drop(columns=['sex']).columns   # —Å–ø–∏—Å–æ–∫ –∫–æ–ª–æ–Ω–æ–∫, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –±—É–¥–µ–º —Ä–∞–±–æ—Ç–∞—å. –ù–∏–∫–∞–∫–æ–≥–æ —Å–µ–∫—Å–∞ :-|

fig, ax = plt.subplots(3, 2, figsize=(18, 18))           # 3 —Å—Ç—Ä–æ–∫–∏ 2 –∫–æ–ª–æ–Ω–∫–∏

for position, column in zip(grid, data_plot_columns):    # –¥–ª—è —è—á–µ–π–∫–∏, –∫–æ–ª–æ–Ω–∫–∏ –≤ zip(—Å–µ—Ç–∫–µ, —Å–ø–∏—Å–∫–µ –∫–æ–ª–æ–Ω–æ–∫)
    sns.histplot(data=data, x=column, ax=ax[position])   # histplot(—Ç–∞–±–ª–∏—Ü–∞, –∫–æ–ª–æ–Ω–∫–∞, ax=ax[—è—á–µ–π–∫–∞])

plt.tight_layout()
plt.show()
```

–ò–Ω–æ–≥–¥–∞ –≤–æ–∑–Ω–∏–∫–∞—é—Ç —Å–∏—Ç—É–∞—Ü–∏–∏, –∫–æ–≥–¥–∞ –º–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –ø—Ä–æ—â–µ –æ—Ç—Ä–∏—Å–æ–≤–∞—Ç—å –≤—ã—à–µ–æ–∑–Ω–∞—á–µ–Ω–Ω—ã–º –æ–±—Ä–∞–∑–æ–º. –ù–æ –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏ —ç—Ç–æ–º –Ω–∞–¥–æ –∏–∑–º–µ–Ω–∏—Ç—å –ø–æ–¥–ø–∏—Å—å –æ—Å–∏, —Ä–∞–∑—Ä—è–¥–Ω–æ—Å—Ç—å, –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è, —Ü–≤–µ—Ç –ø–æ–¥–ø–∏—Å–µ–π, –Ω–∞–∫–æ–Ω–µ—Ü. –ò —Ç–∞–∫–æ–π —Å–ø–æ—Å–æ–± –µ—Å—Ç—å.     


–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Å—Ç—Ä–æ–∫–∏, —Å–Ω–∞–±–∂—ë–Ω–Ω—ã–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è–º–∏ - –æ—Å—Ç–∞–ª—å–Ω–æ–π –∫–æ–¥ –∑–¥–µ—Å—å —Ç–∞–∫–æ–π –∂–µ –∫–∞–∫ –≤—ã—à–µ. 

```python
fig, ax = plt.subplots(3, 2, figsize=(18, 18))

for position, column in zip(grid, data_plot_columns):
    sns.histplot(data=data, x=column, ax=ax[position])

plt.sca(ax[0, 1])                  # –æ–±—Ä–∞—â–µ–Ω–∏–µ –∫ —è—á–µ–π–∫–µ —Å —Å—Ç—Ä–æ–∫–µ 0, –∫–æ–ª–æ–Ω–∫–µ 1
plt.xlabel('–û–°–¢–†–û–í')               # –∑–∞–º–µ–Ω–∞ –ø–æ–¥–ø–∏—Å–∏ –æ—Å–∏ x
plt.sca(ax[2, 1])                  # –æ–±—Ä–∞—â–µ–Ω–∏–µ –∫ —è—á–µ–π–∫–µ —Å —Å—Ç—Ä–æ–∫–µ 2, –∫–æ–ª–æ–Ω–∫–µ 1
plt.xticks([3000, 4500, 6000])      # –ø—Ä–∏—Å–æ–≤–µ–Ω–∏–µ '—Å–≤–æ–∏—Ö' —Ç–∏–∫–æ–≤

plt.suptitle('–ü–∏–Ω–≥–≤–∏–Ω—å–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.tight_layout()
plt.show()
```

–ú–æ–∂–Ω–æ –≤ —Ü–∏–∫–ª–µ –≤ —Å–∞–±–ø–ª–æ—Ç–µ —Ä–∞–∑–º–µ—Å—Ç–∏—Ç—å –∏ —Ä–∞–∑–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏. –ù–∞–ø—Ä–∏–º–µ—Ä, —Å–ª–µ–≤–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∞ —Å–ø—Ä–∞–≤–∞ —Ä–∞—Å—Å–µ—è–Ω–∏–µ. –≠—Ç–æ –¥–∞–∂–µ –ø—Ä–æ—â–µ. –¢—É—Ç –º—ã, –∫–∞–∫ —Ä–∞–∑ –∏—Å–ø–æ–ª—å–∑—É–µ–º –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –ø—Ä—è–º–æ –≤ –∞—Ä–≥—É–º–µ–Ω—Ç–∞—Ö —Ü–∏–∫–ª–∞, –Ω–µ –ø—Ä–∏—Å–≤–∞–∏–≤–∞—è –µ–≥–æ –Ω–∏–∫–∞–∫–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
